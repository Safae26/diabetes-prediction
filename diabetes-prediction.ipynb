### Workflow :

#### Diabetes --> Data Preprocessing --> Train Test Split --> Support Vector Machine Classifier 

#### New Data --> Trained Model --> Diabetic/Non-Diabetic

### Dependencies

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler # Standardize the data to a common range 
from sklearn.model_selection import train_test_split

from sklearn import svm
from sklearn.metrics import accuracy_score

import warnings 
warnings.filterwarnings('ignore')
```

### Data Collection and Analysis

```python
# It is a Female Data
diabetes_dataset = pd.read_csv('/kaggle/input/diabetes-dataset/diabetes.csv') # PIMA Diabetes Dataset 
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
# Statistical measures
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
# In big projects that need neural networks for example, we need more data than this.
# Also, the classes should be equal
```

0 --> Non-Diabetic

1 --> Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Features
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)

# Target
Y = diabetes_dataset['Outcome']
```

```python
X
```

```python
Y
```

### Data Standardization

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X) # Or 2 in 1 using : fit_transform
print(standardized_data)
```

```python
X = standardized_data
```

```python
X
```

```python
Y
```

### Splitting Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(f"Shapes - Original: {X.shape}, Train: {X_train.shape}, Test: {X_test.shape}")
```

### Training the model

```python
# Linear SVM
classifier = svm.SVC(kernel='linear') # SVC : Support Vector Classifier
```

```python
classifier.fit(X_train, Y_train)
```

### Evaluation

```python
# Training Data Accuracy Score :

X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# Testing Data Accuracy Score :

X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', testing_data_accuracy)
```

```python
input_data = [5,166,72,19,175,25.8,0.587,51]

# Changing it to a NumPy array + Reshaping
input_data_array = np.array(input_data).reshape(1, -1)

# standardize the input data
std_data = scaler.transform(input_data_array)

pred = classifier.predict(std_data)

if pred[0] == 1:
    print("This person is diabetic")
else:
    print("This person is not diabetic")
```
